```python
matches = sb.matches(competition_id=11, season_id=27)
matches.head()
```

```python
#Gets the matches

matches[
    [
        'match_id',
        'home_team',
        'away_team',
        'match_date'

    ]
].head()
```

```python
sample_matches = matches.head(10)
sample_matches.shape
```

```python
#We get the events per match 

all_events = []

for match_id in sample_matches['match_id']:

    print(f"Downloading {match_id}")

    match_events = sb.events(match_id=match_id)

    all_events.append(match_events)
```

```python
type(all_events[0])
```

```python
events = pd.concat(
    all_events,
    ignore_index=True
)
```

```python
events.shape
```

```python
events.columns
```

```python
events.head()
```

```python
events.info()
```

```python
events['type_clean'] = (
    events['type']
    .astype(str)
    .str.strip()
)
```

```python
events['type_clean'].value_counts()
```

```python
#Save copy
events_master = events.copy()
```

```python
passes = events[events['type'] == 'Pass']
passes.shape
```

```python
passes_by_player = (
    passes
    .groupby('player')
    .size()
    .reset_index(name='passes')
)

passes_by_player.head()
```

```python
passes_by_player.sort_values(
    by='passes',
    ascending=False
).head(15)
```

```python
shots = events[
    events['type_clean'] == 'Shot'
]
shots.shape
```

```python
shots.head()
```

```python
shots_by_player = (
    shots
    .groupby('player')
    .size()
    .reset_index(name='shots')
)
shots_by_player.head()
```

```python
xg_by_player = (
    shots
    .groupby('player')['shot_statsbomb_xg']
    .sum()
    .reset_index(name='xg')
)
xg_by_player.head(15)
```
