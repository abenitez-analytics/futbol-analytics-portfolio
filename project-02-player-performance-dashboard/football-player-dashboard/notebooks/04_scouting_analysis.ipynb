```python
import pandas as pd
import matplotlib.pyplot as plt 

player_stats = pd.read_csv('../data/cleaned/player_stats_v2.csv')
```

```python
player_stats.info()
```

```python
player_stats.describe()
```

```python
top_attackers = (
    player_stats
    .sort_values(
        by='attacking_score',
        ascending=False
    )
)

top_attackers.head(15)
```

```python
top_pressers = (
    player_stats
    .sort_values(
        by='pressures_per90',
        ascending=False
    )
)

top_pressers.head(15)
```

```python
plt.figure(figsize=(12,8))

plt.scatter(
    player_stats['xg_per90'],
    player_stats['pressures_per90']
)

plt.xlabel('xG per 90')
plt.ylabel('Pressures per 90')
plt.title('La Liga Scouting Map')

plt.show()
```

```python
top_players = player_stats.nlargest(
    10,
    'attacking_score'
)

top_players
```

```python
plt.figure(figsize=(12,8))

plt.scatter(
    player_stats['xg_per90'],
    player_stats['pressures_per90'],
    alpha=0.6
)

for _, row in top_players.iterrows():

    plt.annotate(
        row['player'],
        (
            row['xg_per90'],
            row['pressures_per90']
        )
    )

plt.xlabel('xG per 90')
plt.ylabel('Pressures per 90')
plt.title('La Liga Scouting Map')

plt.savefig(
    '../visuals/la_liga_scouting_map.png',
    dpi=300,
    bbox_inches='tight'
)

plt.show()
```

```python
shortlist = player_stats[
    (player_stats['xg_per90'] >= 0.20) &
    (player_stats['pressures_per90'] >= 5)
]

```

```python
shortlist.sort_values(
    by='xg_per90',
    ascending=False
)

shortlist.shape
```
